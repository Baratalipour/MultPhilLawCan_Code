```python
# file paths
labels_path = '/Users/mahdi/Library/CloudStorage/OneDrive-UniversityofToronto/6DomainsSeparatelyIn4text/labels.txt'
principles_path = '/Users/mahdi/Library/CloudStorage/OneDrive-UniversityofToronto/6DomainsSeparatelyIn4text/principles.txt'
output_path = '/Users/mahdi/Library/CloudStorage/OneDrive-UniversityofToronto/6DomainsSeparatelyIn4text/output_principles.txt'

# read files
with open(labels_path) as f:
  labels = f.read().splitlines()
  
with open(principles_path) as f:
  principles = f.read().splitlines()
  
# print files  
print(labels)
print(principles)

# preprocess text
def preprocess(text):
  text = text.lower()  
  text = text.strip()
  return text

labels = [preprocess(label) for label in labels]  
principles = [preprocess(principle) for principle in principles]

# filter using substring match
filtered = [p for p in principles 
            if any(l in p for l in labels)]
            
# output            
print(filtered)

with open(output_path, 'w') as f:
  for item in filtered:
    f.write(item + '\n')
```
