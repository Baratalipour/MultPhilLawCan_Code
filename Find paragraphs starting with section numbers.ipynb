# Find paragraphs starting with section numbers

```python
import re

# Define the input directory
input_directory = "/Users/mahdi/Library/CloudStorage/OneDrive-UniversityofToronto/6DomainsSeparatelyIn4text/"

# Open and read the content of the file 'ConstitutionAct1867.txt'
with open(input_directory + 'ConstitutionAct1867.txt') as file:
    text = file.read()

# Use regular expressions to filter paragraphs
law_paragraphs = re.findall(r'^\d+[\s.].*$', text, flags=re.MULTILINE)

# Print the filtered paragraphs
for paragraph in law_paragraphs:
    print(paragraph)
```
